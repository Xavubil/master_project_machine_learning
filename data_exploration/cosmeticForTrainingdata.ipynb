# Aufbereiten des Datensatzes fürs training (kleine Kosmetikaufgaben)

```python
import pandas as pd

dfAll = pd.read_csv("allLabeled.csv", encoding="utf-8",sep=",")
dfAll.head()
```

```python
dfAll.drop(inplace=True, columns=["Messtaster","Grenzlehrdorn","Ebene","Bohrung"])
print(dfAll.shape)
```

```python
dfAll.rename(columns={"Messschraube" : "Label"}, inplace=True)
dfAll = dfAll[[c for c in dfAll if c not in ["Label"]] + ["Label"]]
dfAll.head()
```

```python
dfAll.to_csv("trainingdataset.csv", encoding="utf-8",sep=",", index=False)
```
